```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
tesla = pd.read_csv('Historical_Stock_Data/TSLAStock.csv')
tesla
```

```python
tesla.columns = ['Date', 'Close', 'Volume', 'Open', 'High', 'Low']
```

```python
fig = plt.figure(figsize = (10, 5))
plt.plot(tesla['Date'], tesla['Close'], color ='blue') 
plt.xlabel("Date") 
plt.xticks(rotation=90)
plt.ylabel("Closing Price") 
plt.title("Closing Price per Day") 
plt.show()
```
